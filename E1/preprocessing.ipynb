## RE

```python
def load_data(file):
    with open(file, 'r') as f:
        data = json.load(f)
    return data
```

```python
path_ita = '/Users/joelpardo/Desktop/PAQUITA/Italian/preprocessed/relations_dev.txt'
path_spa = '/Users/joelpardo/Desktop/PAQUITA/Spanish/preprocessed/relations_dev.txt'

path_ita1 = '/Users/joelpardo/Desktop/PAQUITA/Italian/preprocessed/relations_test.txt'
path_spa1 = '/Users/joelpardo/Desktop/PAQUITA/Spanish/preprocessed/relations_test.txt'

path_ita2 = '/Users/joelpardo/Desktop/PAQUITA/Italian/preprocessed/relations_training.txt'
path_spa2 = '/Users/joelpardo/Desktop/PAQUITA/Spanish/preprocessed/relations_training.txt'

ita = load_data(path_ita)
spa = load_data(path_spa)

ita1 = load_data(path_ita1)
spa1 = load_data(path_spa1)

ita2 = load_data(path_ita2)
spa2 = load_data(path_spa2)

pprint(len(ita))
pprint(len(spa))

pprint(len(ita1))
pprint(len(spa1))

pprint(len(ita2))
pprint(len(spa2))
```

```python
def shuffle_data(lang1, lang2):
    data1 = load_data(lang1)
    data2 = load_data(lang2)
    
    combined_data = data1 + data2
    random.shuffle(combined_data)

    return combined_data
```

```python
# Prueba la funcin
shuffle_data(path_ita, path_spa)  # reemplaza con los nombres correctos de tus archivos
```

```python
def save_data(filename, data):
    with open(filename, 'w') as f:
        json.dump(data, f, ensure_ascii=False, indent=4)
```

```python
# Cargar y barajar los datos
shuffled_data = shuffle_data(path_ita, path_spa)
shuffled_data1 = shuffle_data(path_ita1, path_spa1)
shuffled_data2 = shuffle_data(path_ita2, path_spa2)

# Guardar los datos barajados
save_data('/Users/joelpardo/Desktop/PAQUITA/E1/preprocessed/relations_dev.txt', shuffled_data)
save_data('/Users/joelpardo/Desktop/PAQUITA/E1/preprocessed/relations_test.txt', shuffled_data1)
save_data('/Users/joelpardo/Desktop/PAQUITA/E1/preprocessed/relations_training.txt', shuffled_data2)
```

## NER

```python
def load_data(file):
    with open(file, 'r') as f:
        doc = []
        sentence = []
        for line in f:
            line = line.strip()
            if line == '-DOCSTART- -X- O O':  # new sentence
                if sentence:
                    doc.append(sentence)
                    sentence = []
            else:
                sentence.append(line)
        # Append last sentence if it exists
        if sentence:
            doc.append(sentence)
    return doc
```

```python
path_ita = '/Users/joelpardo/Desktop/PAQUITA/Italian/preprocessed/NER_DEV.tsv'
path_spa = '/Users/joelpardo/Desktop/PAQUITA/Spanish/preprocessed/NER_DEV.tsv'

path_ita1 = '/Users/joelpardo/Desktop/PAQUITA/Italian/preprocessed/NER_TEST.tsv'
path_spa1 = '/Users/joelpardo/Desktop/PAQUITA/Spanish/preprocessed/NER_TEST.tsv'

path_ita2 = '/Users/joelpardo/Desktop/PAQUITA/Italian/preprocessed/NER_training.tsv'
path_spa2 = '/Users/joelpardo/Desktop/PAQUITA/Spanish/preprocessed/NER_training.tsv'

ita = load_data(path_ita)
spa = load_data(path_spa)

ita1 = load_data(path_ita1)
spa1 = load_data(path_spa1)

ita2 = load_data(path_ita2)
spa2 = load_data(path_spa2)

pprint(len(ita))
pprint(len(spa))

pprint(len(ita1))
pprint(len(spa1))

pprint(len(ita2))
pprint(len(spa2))
```

```python
def shuffle_data(lang1, lang2):
    data1 = load_data(lang1)
    data2 = load_data(lang2)
    
    combined_data = data1 + data2
    
    random.shuffle(combined_data)

    return combined_data
```

```python
# Prueba la funcin
shuffle_data(path_ita, path_spa)  # reemplaza con los nombres correctos de tus archivos
```

```python
def save_data(filename, data):
    with open(filename, 'w') as f:
        for sentence in data:
            f.write('-DOCSTART- -X- O O\n')  # sentence separator
            for line in sentence:
                f.write(f'{line}\n')  # write each line of the sentence
```

```python
# Cargar y barajar los datos
shuffled_data = shuffle_data(path_ita, path_spa)
shuffled_data1 = shuffle_data(path_ita1, path_spa1)
shuffled_data2 = shuffle_data(path_ita2, path_spa2)

# Guardar los datos barajados
save_data('/Users/joelpardo/Desktop/PAQUITA/E1/preprocessed/NER_DEV.tsv', shuffled_data)
save_data('/Users/joelpardo/Desktop/PAQUITA/E1/preprocessed/NER_TEST.tsv', shuffled_data1)
save_data('/Users/joelpardo/Desktop/PAQUITA/E1/preprocessed/NER_training.tsv', shuffled_data2)
```
